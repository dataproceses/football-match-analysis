### 1. Importing Esential Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Dataset Import and Preparation

```python
#loading data
df = pd.read_csv('season-1718.csv')
```

```python
#display five rows
df.head()
```

```python
#no of rows and columns
df.shape
```

```python
# data overview
df.info()
```

```python
# stastical overview
df.describe()
```

```python
# finding null value if any
df.isnull().sum().sum()
```

```python
# finding dublicated value in any
df.duplicated().sum()
```

```python
#changing Data Types
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
df['Total_Goals'] = df['FTHG'] + df['FTAG']
df['Total_Half_Goals'] = df['HTHG'] + df['HTAG']
df['Total_Shots'] = df['HS'] + df['HST']
df['Total_Corners'] = df['HC'] + df['AC']
df['Total_Fouls'] = df['HF'] + df['AF']
df['Total_Yellow_Cards'] = df['HY'] + df['AY']
df['Total_Red_Cards'] = df['HR'] + df['AR']
```

# 3. Explatory Data Analysis

#### 3.1 Monthly Match Statistics Overview

```python
Monthly_Total_Goals = df.groupby('Month').agg(
    Total_Goals = ('Total_Goals', 'sum'),
    Average_Goals = ('Total_Goals', 'mean')
).reset_index()

Monthly_Half_Goals = df.groupby('Month').agg(
    Total_Half_Goals = ('Total_Half_Goals', 'sum'),
    Average_Half_Goals = ('Total_Half_Goals', 'mean')
).reset_index()

Monthly_total_shots = df.groupby('Month').agg(
    Total_Shots = ('Total_Shots', 'sum'),
    Average_Shots = ('Total_Shots', 'mean')
).reset_index()

Monthly_Corners = df.groupby('Month').agg(
    Total_Corners = ('Total_Corners', 'sum'),
    Average_Corners = ('Total_Corners', 'mean')
).reset_index()

Monthly_Fouls = df.groupby('Month').agg(
    Total_Fouls = ('Total_Fouls', 'sum'),
    Average_Fouls = ('Total_Fouls', 'mean')
).reset_index()

Monthly_Yellow_Cards = df.groupby('Month').agg(
    Total_Yellow_Cards = ('Total_Yellow_Cards', 'sum'),
    Average_Yellow_Cards = ('Total_Yellow_Cards', 'mean')
).reset_index()

Monthly_Red_Cards = df.groupby('Month').agg(
    Total_Red_Cards = ('Total_Red_Cards', 'sum'),
    Average_Red_Cards = ('Total_Red_Cards', 'mean')
).reset_index()

plt.figure(figsize=(17,17))

plt.subplot(3,3,1)
sns.lineplot(data = Monthly_Total_Goals, x = 'Month', y = 'Total_Goals')
plt.xlabel('Month')
plt.ylabel('Total Goals')
plt.title('Monthly Total Goals')

plt.subplot(3,3,2)
sns.lineplot(data = Monthly_Half_Goals, x = 'Month', y = 'Total_Half_Goals')
plt.xlabel('Month')
plt.ylabel('Total Half Goals')
plt.title('Monthly Total Half Goals')

plt.subplot(3,3,3)
sns.lineplot(data = Monthly_total_shots, x = 'Month', y = 'Total_Shots')
plt.xlabel('Month')
plt.ylabel('Total Shots')
plt.title('Monthly Total Shots')

plt.subplot(3,3,4)
sns.lineplot(data = Monthly_Corners, x = 'Month', y = 'Total_Corners')
plt.xlabel('Month')
plt.ylabel('Total Corners')
plt.title('Monthly Total Corners Per ')

plt.subplot(3,3,5)
sns.lineplot(data = Monthly_Fouls, x = 'Month', y = 'Total_Fouls')
plt.xlabel('Month')
plt.ylabel('Total Fouls')
plt.title('Monthly Total Fouls')

plt.subplot(3,3,6)
sns.lineplot(data = Monthly_Yellow_Cards, x = 'Month', y = 'Total_Yellow_Cards')
plt.xlabel('Month')
plt.ylabel('Total Yellow Card')
plt.title('Monthly Total Yellow Cards')

plt.subplot(3,3,7)
sns.lineplot(data = Monthly_Red_Cards, x = 'Month', y = 'Total_Red_Cards')
plt.xlabel('Month')
plt.ylabel('Total Red Cards')
plt.title('Monthly Total Red Cards')

plt.suptitle("Monthly Match Statistics Overview")
plt.tight_layout()
plt.show()
```

#### 3.2 Wins And Drow Per Team

```python
plt.figure(figsize=(17,15))
plt.subplot(2,2,1)
home_team_win = df[df['FTR'] == "H"]['HomeTeam'].value_counts().reset_index()
sns.barplot(y='HomeTeam', x='count', data=home_team_win, orient='h',color='#32876C')
plt.title("Home Team Win")

plt.subplot(2,2,2)
away_team_win = df[df['FTR'] == "A"]['AwayTeam'].value_counts().reset_index()
sns.barplot(y='AwayTeam', x='count', data=away_team_win, orient='h', color='#693287')
plt.title("Away Team Win")

plt.subplot(2,2,3)
home_team_drawn = df[df['FTR'] == "D"]['HomeTeam'].value_counts().reset_index()
sns.barplot(y='HomeTeam', x='count', data=home_team_drawn, orient='h',color='#DDA416')
plt.title("Home Team Drawn")

plt.subplot(2,2,4)
away_team_drawn = df[df['FTR'] == "D"]['AwayTeam'].value_counts().reset_index()
sns.barplot(y='AwayTeam', x='count', data=away_team_drawn, orient='h',color='#AEC416')
plt.title("Away Team Drawn")

plt.tight_layout()
plt.show()
```

#### 3.3 Team Goals

```python
home_team_goals = df.groupby('HomeTeam')['FTHG'].sum().sort_values(ascending=False).reset_index()
away_team_goals = df.groupby('AwayTeam')['FTAG'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.barplot(y='HomeTeam', x='FTHG', data=home_team_goals, orient='h',color='#32876C')
plt.title("Home Team Total Goals")

plt.subplot(1,2,2)
sns.barplot(y='AwayTeam', x='FTAG', data=away_team_goals, orient='h', color='#693287')
plt.title("Away Team Total Goals")

plt.suptitle('Team Goals')
plt.tight_layout()
plt.show()
```

#### 3.4 Team Shots

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a nice style
sns.set_theme(style="whitegrid")

# Prepare data
home_teams_shots = df.groupby('HomeTeam')['HS'].sum().sort_values(ascending=False).reset_index()
home_teams_shots_ontarget = df.groupby('HomeTeam')['HST'].sum().sort_values(ascending=False).reset_index()
away_teams_shots = df.groupby('AwayTeam')['HS'].sum().sort_values(ascending=False).reset_index()
away_teams_shots_ontarget = df.groupby('AwayTeam')['HST'].sum().sort_values(ascending=False).reset_index()

# Colors
main_color = "#1f77b4"   # professional blue
target_color = "#ff7f0e" # clean orange
bg_color = "#2ca02c"     # green for away
bg_target_color = "#d62728" # red for away on target

# Plot
plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
sns.barplot(data=home_teams_shots, x='HS', y='HomeTeam', color=main_color)
plt.xlabel('Total Shots')
plt.ylabel('Home Team')
plt.title('Home Teams - Total Shots')

plt.subplot(2, 2, 2)
sns.barplot(data=home_teams_shots_ontarget, x='HST', y='HomeTeam', color=target_color)
plt.xlabel('Shots on Target')
plt.ylabel('Home Team')
plt.title('Home Teams - Shots On Target')

plt.subplot(2, 2, 3)
sns.barplot(data=away_teams_shots, x='HS', y='AwayTeam', color=bg_color)
plt.xlabel('Total Shots')
plt.ylabel('Away Team')
plt.title('Away Teams - Total Shots')

plt.subplot(2, 2, 4)
sns.barplot(data=away_teams_shots_ontarget, x='HST', y='AwayTeam', color=bg_target_color)
plt.xlabel('Shots on Target')
plt.ylabel('Away Team')
plt.title('Away Teams - Shots On Target')

# Super Title
plt.suptitle('Number Of Shots Per Team (Home & Away)', fontsize=20, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### 3.5 Team Corners

```python
home_team_corners = df.groupby('HomeTeam')['HC'].sum().sort_values(ascending=False).reset_index()
away_team_corners = df.groupby('AwayTeam')['AC'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.barplot(y='HomeTeam', x='HC', data=home_team_corners, orient='h',color='#32876C')
plt.title("Home Team Total Corners")

plt.subplot(1,2,2)
sns.barplot(y='AwayTeam', x='AC', data=away_team_corners, orient='h', color='#693287')
plt.title("Away Team Total Corners")

plt.suptitle('Team Corners')
plt.tight_layout()
plt.show()
```

#### 3.6 Team Fouls

```python
home_team_fouls = df.groupby('HomeTeam')['HF'].sum().sort_values(ascending=False).reset_index()
away_team_fouls = df.groupby('AwayTeam')['AF'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.barplot(y='HomeTeam', x='HF', data=home_team_fouls, orient='h',color='#32876C')
plt.title("Home Team Total Fouls")

plt.subplot(1,2,2)
sns.barplot(y='AwayTeam', x='AF', data=away_team_fouls, orient='h', color='#693287')
plt.title("Away Team Total Fouls")

plt.suptitle('Team Fouls')
plt.tight_layout()
plt.show()
```

#### 3.7 Team Cards

```python
home_team_yellow_cards = df.groupby('HomeTeam')['HY'].sum().sort_values(ascending=False).reset_index()
away_team_yellow_cards = df.groupby('AwayTeam')['AY'].sum().sort_values(ascending=False).reset_index()
home_team_red_cards = df.groupby('HomeTeam')['HR'].sum().sort_values(ascending=False).reset_index()
away_team_red_cards = df.groupby('AwayTeam')['AR'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.barplot(y='HomeTeam', x='HY', data=home_team_yellow_cards, orient='h',color='Yellow')
plt.title("Home Team Total Yellow Cards")

plt.subplot(2,2,2)
sns.barplot(y='AwayTeam', x='AY', data=away_team_yellow_cards, orient='h', color='Yellow')
plt.title("Away Team Total Yellow Cards")

plt.subplot(2,2,3)
sns.barplot(y='HomeTeam', x='HR', data=home_team_red_cards , orient='h', color='red')
plt.title("Home Team Total Red Cards")

plt.subplot(2,2,4)
sns.barplot(y='AwayTeam', x='AR', data=away_team_red_cards , orient='h', color='red')
plt.title("Away Team Total Red Cards")


plt.suptitle('Team Cards')
plt.tight_layout()
plt.show()
```

```python

# Sum totals
totals = df[['Total_Goals', 'Total_Half_Goals', 'Total_Shots', 'Total_Corners', 'Total_Fouls', 'Total_Yellow_Cards', 'Total_Red_Cards']].sum()

# Sort from largest to smallest
totals = totals.sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(totals.index, totals.values, color='royalblue')

# Add value labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, int(yval), ha='center', va='bottom', fontsize=10)

# Titles and labels
plt.title('Total Match Statistics Distribution', fontsize=18, fontweight='bold')
plt.ylabel('Total Count', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
